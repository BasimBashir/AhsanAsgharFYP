```python

from scipy import stats
import numpy as np
 
z = np.abs(stats.zscore(data))
print("Printing Z_data:\n", z)
    
print('\n\n')
threshold = 3             #Setting the outlier removal index to 3
 
# Position of the outlier
print("Indexes where outliers are found: \n", np.where(z > 3))
    
print('\n\n')
    
dt = data[(z < 3).all(axis=1)] #Storing to a new variable i.e., without outliers use dt from now on as a dataset
print("Before outlier Removal: ", data.shape, "\nAfter Outlier Removal: ", dt.shape)
```

```python
X = dt.iloc[:, :3].values
X
```

```python
y = dt.iloc[:, -1].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# transform data
scaler.fit_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
r2_score = lr.score(X_test, y_test)
print(r2_score*100,'%')
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
```

```python
y_pred2 = dt.predict(X_test)
y_pred2
```

```python
dt_score = dt.score(X_test, y_test)
print(dt_score*100,'%')
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(max_depth=7, random_state=1, n_estimators=100, n_jobs=-1)
model.fit(X_train, y_train)
```

```python
y_pred3 = model.predict(X_test)
y_pred3
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = { 
            "n_estimators"      : [10,20,30,50,100,200,500],
            "max_depth"      : [1,3,5,7,9,11,13,15,17],
            "max_features"      : ["auto", "sqrt", "log2"],
            }

grid = GridSearchCV(model, param_grid, n_jobs=-1, cv=5)

grid.fit(X_train, y_train)

grid.best_score_ , grid.best_params_
```

```python
model = RandomForestRegressor(max_depth=11, random_state=1, n_estimators=20, n_jobs=-1)
model.fit(X_train, y_train)
print((model.score(X_test, y_test))*100, " %")
```
